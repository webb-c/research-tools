```python
title = input("title: ")
```

```python
sortby = "cit/year"         # @param ["Citations", "cit/year"] {type:"string"}
num_results = 50              
start_year = 2022            
end_year = None              
```

```python
cmd = f"python getcit.py --title '{title}' --sortby '{sortby}' --nresults {num_results} --csvpath results/"

if start_year:
    cmd += f" --startyear {start_year}"

if end_year:
    cmd += f" --endyear {end_year}"


# Output the constructed command for review
print("Constructed command:", cmd)
```
